Bacmman can read nd2 (Nikon files) directly, but not vsi (Olympus). 

If you use Olympuys as us, we need to convert data first. You can do this by hand in Fiji, but here we will use a script to batch convert.

First we need to install the bioformats library.

First make sure you have a new Java JDK installed [see here for instructions](https://www.oracle.com/java/technologies/javase-downloads.html)

```bash
conda activate i2i_p2_env
pip install python-bioformats
```
